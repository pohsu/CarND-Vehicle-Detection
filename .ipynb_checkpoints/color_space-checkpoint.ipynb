```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import glob
import pickle
from utility import *

%matplotlib inline
%load_ext autoreload
%autoreload 2
```

```python
# Read a color image
colornames = glob.glob('color_images/*.png')
img1 = cv2.imread(colornames[0])
img2 = cv2.imread(colornames[4])

# Convert subsampled image to desired color space(s)
img1_color = {}
img1_color['RGB'] = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)  # OpenCV uses BGR, matplotlib likes RGB
img1_color['HSV']  = cv2.cvtColor(img1, cv2.COLOR_BGR2HSV)
img1_color['HLS']  = cv2.cvtColor(img1, cv2.COLOR_BGR2HLS)
img1_color['YUV']  = cv2.cvtColor(img1, cv2.COLOR_BGR2YUV)
img1_color['Yrb']  = cv2.cvtColor(img1, cv2.COLOR_BGR2YCrCb)
img1_rgb = img1_color['RGB'] / 255.  # scaled to [0, 1], only for plotting

# Convert subsampled image to desired color space(s)
img2_color = {}
img2_color['RGB'] = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)  # OpenCV uses BGR, matplotlib likes RGB
img2_color['HSV']  = cv2.cvtColor(img2, cv2.COLOR_BGR2HSV)
img2_color['HLS']  = cv2.cvtColor(img2, cv2.COLOR_BGR2HLS)
img2_color['YUV']  = cv2.cvtColor(img2, cv2.COLOR_BGR2YUV)
img2_color['Yrb']  = cv2.cvtColor(img2, cv2.COLOR_BGR2YCrCb)
img2_rgb = img2_color['RGB'] / 255.  # scaled to [0, 1], only for plotting

# %matplotlib inline
# Plot and show
plt.subplot(121)
plt.imshow(img1_color['RGB'])
plt.subplot(122)
plt.imshow(img2_color['RGB'])
plt.show()
```

```python
%matplotlib notebook
idx = 'HSV'
plot3d(img1_color[idx], img1_rgb, axis_labels=list(idx))
plot3d(img2_color[idx], img2_rgb, axis_labels=list(idx))
plt.show()
```
